```python
from Invtsim import *
import numpy as np
import pandas as pd
from scipy.stats import norm
```

```python
path1  = 'D:/Jaypy/Semester 1, 2025/ExtensionHF/JayCode/Reconciliation/'
path2  = 'D:/Jaypy/Semester 1, 2025/ExtensionHF/JayCode/Data/'

del ets, test

def loop(fcst, truth, a = 0.95, df_list = [], length: int = 28, n = None):
    if n == None:
        n = int(len(fcst)/length)
    from tqdm import tqdm
    for i in tqdm(range(n)):
        fcst_1  = fcst[i*length:(i+1)*length]
        truth_1 = truth[i*length:(i+1)*length]
        ds = InvtSim(j=i, fcst = fcst_1, truth = truth_1, alpha = a)
        res = ds.ob_all_t()
        df_list.append(res.reset_index(drop = True))
    #results = pd.concat(res)
    print("-----------DOWN!------------")
    return df_list
```

## AutoETS base:

```python
ets     = pd.read_pickle(f"{path1}AutoETSfcst.pkl")
test    = pd.read_pickle(f"{path2}test.pkl")
etsfcst = ets['AutoETS']
truth   = test["y"]

asl_99 = loop(fcst = etsfcst, truth = truth, a = 0.99,n = 42840)
a99 = pd.concat(asl_99).reset_index(drop = True)
a99.to_pickle("ETS_base_99.pkl")
a99
```

```python
asl_95 = loop(fcst = etsfcst, truth = truth, a = 0.95,n = 42840)
a95 = pd.concat(asl_95).reset_index(drop = True)
a95.to_pickle("ETS_base_95.pkl")
a95
```

```python
asl_90 = loop(fcst = etsfcst, truth = truth, a = 0.90,n = 42840)
a90 = pd.concat(asl_90).reset_index(drop = True)
a90.to_pickle("ETS_base_90.pkl")
a90
```

```python
lgbbase = pd.read_pickle(f"{path1}LGBMfcst.pkl")
test    = pd.read_pickle(f"{path2}test.pkl")
lgbfcst = lgbbase['LGBMRegressor']
truth   = test["y"]

asl_99 = loop(fcst = lgbfcst, truth = truth, a = 0.99,n = 42840)
a99 = pd.concat(asl_99).reset_index(drop = True)
a99.to_pickle("LGBM_base_99.pkl")

asl_95 = loop(fcst = lgbfcst, truth = truth, a = 0.95,n = 42840)
a95 = pd.concat(asl_95).reset_index(drop = True)
a95.to_pickle("LGBM_base_95.pkl")

asl_90 = loop(fcst = lgbfcst, truth = truth, a = 0.90,n = 42840)
a90 = pd.concat(asl_90).reset_index(drop = True)
a90.to_pickle("LGBM_base_90.pkl")
```
