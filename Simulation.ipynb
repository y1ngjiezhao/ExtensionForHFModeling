```python
from Invtsim import *
import numpy as np
import pandas as pd
from scipy.stats import norm
import warnings
warnings.simplefilter("ignore")
```

```python
path1  = 'D:/Jaypy/Semester 1, 2025/ExtensionHF/JayCode/Reconciliation/'
path2  = 'D:/Jaypy/Semester 1, 2025/ExtensionHF/JayCode/Data/'

def loop(fcst, truth, a = 0.95, df_list = [], length: int = 28, n = None):
    if n == None:
        n = int(len(fcst)/length)
    from tqdm import tqdm
    for i in tqdm(range(n)):
        fcst_1  = fcst[i*length:(i+1)*length]
        truth_1 = truth[i*length:(i+1)*length]
        ds = InvtSim(j=i, fcst = fcst_1, truth = truth_1, alpha = a)
        res = ds.ob_all_t()
        df_list.append(res.reset_index(drop = True))
    #results = pd.concat(res)
    print("-----------DOWN!------------")
    return df_list
```

## AutoETS base:

```python
ets     = pd.read_pickle(f"{path1}AutoETSfcst.pkl")
test    = pd.read_pickle(f"{path2}test.pkl")
etsfcst = ets['AutoETS']
truth   = test["y"]
```

```python
ets_sl_99 = loop(fcst = etsfcst, truth = truth, a = 0.99,n = 42840)
ets_99 = pd.concat(ets_sl_99).reset_index(drop = True)
ets_99.to_pickle("ETS_base_99.pkl")
print(f"{len(ets_99)}\n{ets_99}")
```

```python
ets_sl_95 = loop(fcst = etsfcst,df_list = [], truth = truth, a = 0.95,n = 42840)
ets_95 = pd.concat(ets_sl_95).reset_index(drop = True)
ets_95.to_pickle("ETS_base_95.pkl")
print(f"{len(ets_95)}\n{ets_95}")
```

```python
ets_sl_90 = loop(fcst = etsfcst,df_list = [], truth = truth, a = 0.90,n = 42840)
ets_90 = pd.concat(ets_sl_90).reset_index(drop = True)
ets_90.to_pickle("ETS_base_90.pkl")
print(f"{len(ets_90)}\n{ets_90}")
```

```python
l1 = ets_99[['backlog']]
print(f"99%:{len(np.where(l1==0)[0])/len(l1)}")

l1 = ets_95[['backlog']]
print(f"95%:{len(np.where(l1==0)[0])/len(l1)}")

l1 = ets_90[['backlog']]
print(f"90%:{len(np.where(l1==0)[0])/len(l1)}")
```
