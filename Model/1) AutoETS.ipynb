```python
%%time
from tqdm import tqdm
sf = StatsForecast(
    models=[AutoETS()],  # periodicity with 365 (whole year
    freq=1,  # 
    n_jobs=61  # all cpu core : -1
)
forecasts = []
fitted    = []
train = data[data['ds']<=1913]
fcst = sf.forecast(df=train,h=1,fitted = True)
ets_fitted = sf.forecast_fitted_values()
#forecasts = fcst
#fitted.append(ets_fitted)
```

```python
%%time
sf = StatsForecast(
    models=[AutoETS()],  # periodicity with 365 (whole year
    freq=1,  # 
    n_jobs=61  # all cpu core : -1
)
forecasts = []
fitted    = []
for i in tqdm(range(28)):
    train = data[data['ds']<=1913+i]
    fcst = sf.forecast(df=train,h=1)
    #ets_fitted = sf.forecast_fitted_values()
    forecasts.append(fcst)
    #fitted.append(ets_fitted)
```
