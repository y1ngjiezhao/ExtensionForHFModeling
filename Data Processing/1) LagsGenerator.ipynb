```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_pickle("processed_data.pkl")
```

```python
def _get_lags(df:pd.DataFrame, col = "y", gap1:int = 1941, gap2:int = 7, n:int=42840):
    from tqdm import tqdm
    return pd.concat([pd.DataFrame({f"{'y'}_lags_{i+1}": df[col][j*gap1:(j+1)*gap1].shift(i+1) for i in range(gap2)}) for j in tqdm(range(n))])
```

```python
lags = _get_lags(df = data)
```

```python
%%time
pd.concat([data,lags],axis = 1).to_pickle("final_processed_data.pkl")
```
