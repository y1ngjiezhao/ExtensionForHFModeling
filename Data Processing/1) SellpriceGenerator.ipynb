```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_pickle("data_with_uid.pkl")
dff = df[["unique_id","sell_price"]]
```

```python
def sell_price(df):
    df['price_std'] = (df['sell_price'] - df['sell_price'].mean()) / df['sell_price'].std()
    df['price_max'] = df['sell_price'].rolling(window=30, min_periods=1).max()  
    df['price_min'] = df['sell_price'].rolling(window=30, min_periods=1).min()  
    df['price_mean'] = df['sell_price'].rolling(window=30, min_periods=1).mean()  
    df['price_std_dev'] = df['sell_price'].rolling(window=30, min_periods=1).std()  
    df['price_momentum_1d'] = df['sell_price'] / df['sell_price'].shift(1) - 1  
    df['price_momentum_1m'] = df['sell_price'] / df['sell_price'].shift(30) - 1  
    df['price_momentum_1y'] = df['sell_price'] / df['sell_price'].shift(365) - 1  
    df.fillna(0, inplace=True)
    return df
```

```python
from tqdm import tqdm
import warnings
warnings.filterwarnings("ignore")
dff_l = []
for i in tqdm(range(42840)):
    dff_l.append(sell_price(dff.iloc[i*1941:(i+1)*1941,:]))
```

```python
len(dff_l)
```

```python
price_data = pd.concat(dff_l)
```

```python
price_data = price_data.reset_index(drop=True)
price_data.to_pickle("price_data.pkl")
```
